# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Load data

```python
data = pd.read_csv("creditcard.csv")
data.head()
```

```python
data.shape
```

Class distribution

```python
class_counts = data['Class'].value_counts()
print("\nClass Distribution:\n", class_counts)
```

Percentage of each class

```python
class_percent = class_counts / len(data) * 100
print("\nClass Percentage:\n", class_percent)
```

Visualization

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Class', data=data, color='blue')
plt.title("Class Distribution: Normal (0) vs Fraud (1)")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(class_counts, labels=['Normal','Fraud'], autopct='%1.4f%%', colors=['pink','salmon'], startangle=90)
plt.title("Class Distribution (Pie Chart)")
plt.show()
```

Heat Map

```python
plt.figure(figsize=(12,10))
corr = data.corr()
sns.heatmap(corr, cmap='coolwarm', annot=False)
plt.title("Correlation Heatmap of Features")
plt.show()
```

Transaction Amount

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.histplot(data['Amount'], bins=50, kde=True, color='purple')
plt.title("Transaction Amount Distribution")
plt.subplot(1,2,2)
sns.boxplot(x='Class', y='Amount', data=data, color='lightgreen')
plt.title("Transaction Amount by Class")
plt.xlabel("Class")
plt.ylabel("Amount")
plt.show()
```

Transaction Time

```python
plt.figure(figsize=(8,5))
sns.histplot(data['Time'], bins=50, kde=True, color='green')
plt.title("Transaction Time Distribution")
plt.xlabel("Time (seconds)")
plt.ylabel("Count")
plt.show()
```

Fraud vs Normal Amount

```python
fraud = data[data['Class']==1]
normal = data[data['Class']==0]

print("\nTransaction Amount Summary:")
print("Normal Transactions:")
print(normal['Amount'].describe())
print("\nFraud Transactions:")
print(fraud['Amount'].describe())
```

Fraud transactions are not large; many are small amounts
