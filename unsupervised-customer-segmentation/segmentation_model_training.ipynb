# Segmentation Model Training

# Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

Load data

```python
data = pd.read_csv("Mall_Customers.csv")
```

Features

```python
X = data[['Annual Income (k$)', 'Spending Score (1-100)']]
```

# KMeans **Model**

```python
k = 3
kmeans = KMeans(n_clusters=k, random_state=42)
data['Cluster'] = kmeans.fit_predict(X)
```

# Cluster Visualization

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)',
                hue='Cluster', palette='viridis', data=data, s=50)
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)',
                hue='Cluster', palette='viridis', data=data, s=50)


plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
            s=200, c='black', marker='X')

plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer Segmentation with KMeans (3 Clusters)')
plt.show()
```

```python
for i in range(k):
    cluster_data = data[data['Cluster'] == i]
    print(f"Cluster {i}:")
    print(f" - Number of customers: {len(cluster_data)}")
    print(f" - Avg Income: {cluster_data['Annual Income (k$)'].mean():.2f}")
    print(f" - Avg Spending Score: {cluster_data['Spending Score (1-100)'].mean():.2f}\n")
```

The mall customers form three groups. Cluster 0: high income, low spending (Big Savers). Cluster 1: high income, high spending (Premium Customers). Cluster 2: medium income, medium spending (Average Customers). This segmentation helps the marketing team target each group effectively.
